```python
import os
import json
import glob
```

```python
annotation_dir = "/DATASETS/PCB/Annotations/IC"
annots = glob.glob(os.path.join(annotation_dir, '*.txt'))

image_dir = "/libraries/Mask_RCNN/datasets/pcbdslr/val"
img_files = glob.glob(os.path.join(image_dir, '*.jpg'))

```

```python
fp = open(os.path.join(image_dir, 'IC_annotation_vgg_format.json'), 'w')

annotations = {}
for i, fi in enumerate(img_files):
    entry = {}
    img = os.path.join(annotation_dir, os.path.basename(fi)[:-3] + 'txt')
    if not os.path.exists(img):
        continue
    entry['filename'] = os.path.basename(fi)
    entry['size'] = os.path.getsize(fi)
    entry['file_attributes'] = {}
    entry['fileref'] = ""
    entry['base64_img_data'] = ""
    entry['regions'] = {}
    with open(img, 'r') as f:
        lines = [l.strip().split() for l in f]
    for j, l in enumerate(lines):
        entry['regions'][str(j)] = {}
        entry['regions'][str(j)]['region_attributes'] = {}
        entry['regions'][str(j)]['shape_attributes'] = {}
        entry['regions'][str(j)]['shape_attributes']['name'] = 'polygon'
        ptsx = [int(l[1]), int(l[1])+int(l[3]), int(l[1])+int(l[3]), int(l[1])]
        ptsy = [int(l[2]), int(l[2]), int(l[2])+int(l[4]), int(l[2])+int(l[4])]
        entry['regions'][str(j)]['shape_attributes']['all_points_x'] = ptsx
        entry['regions'][str(j)]['shape_attributes']['all_points_y'] = ptsy
        #entry['regions'][str(j)]['shape_attributes']['x'] = int(l[1])
        #entry['regions'][str(j)]['shape_attributes']['y'] = int(l[2])
        #entry['regions'][str(j)]['shape_attributes']['width'] = int(l[3])
        #entry['regions'][str(j)]['shape_attributes']['height'] = int(l[4])
#        print(entry['regions'][str(j)])
    for e in entry['regions']:
        print(entry['regions'][e])
    key = "%s%d"%(entry['filename'], entry['size'])
    annotations[key] = entry
    
json.dump(annotations, fp, indent=4)
```
